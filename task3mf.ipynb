```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime
```

```python
data= pd.read_csv("/content/householdtask3.csv")
display (data.head(15))
```

```python
plt.scatter(data['year'],data['own'])
plt.title("scatter plot")
plt.xlabel("year")
plt.ylabel("own")
plt.show()
```

```python
plt.plot(data['year'])
plt.plot(data['own'])
plt.title("line plot")
plt.xlabel("year")
plt.ylabel("own")
plt.show()
```

```python
plt.bar(data['year'],data['own'])
plt.title("bar plot")
plt.xlabel("year")
plt.ylabel("own")
plt.show()
```
