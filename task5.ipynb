```python
df.target.value_counts()
```

```python
df.target.value_counts().plot(kind="bar" , color=["red","green"]);
plt.title("heart disease valuse")
plt.xlabel("1 = no heart disease , 0 = heart disease")
plt.ylabel("amount")
plt.show()
```

```python
df.target.value_counts().plot(kind="pie" , figsize=(8,7));
plt.legend(["disease", "no disease"]);
plt.show()
```

```python
df.sex.value_counts()
```

```python
df.sex.value_counts().plot(kind='pie' , figsize =(8,6));
plt.title("male female ratio")
plt.legend(['male','female']);
plt.show()
```

```python
pd.crosstab(df.target,df.sex)
```

```python
sns.countplot(x= 'target', data=df, hue='sex');
plt.title('heart disease frequency for sex')
plt.xlabel('0 = no disease, 1 = disease')

plt.show()
```

```python
df.cp.value_counts().plot(kind='bar', color = ['salmon', 'lightskyblue', 'springgreen','khaki']);
plt.title("chest pain type vs counts")
```

```python
pd.crosstab(df.sex,df.cp)
```

```python
pd.crosstab(df.sex,df.cp).plot(kind='bar', figsize=(10,6), color = ['salmon', 'lightskyblue', 'red','khaki']);
plt.title("type of chest pain for sex")
plt.xlabel("0 = female, 1 = male")
```
